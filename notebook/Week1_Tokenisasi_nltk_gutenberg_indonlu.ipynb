#Tokenisasi Gutenberg


```python
import nltk
from nltk.tokenize import word_tokenize
from collections import Counter
```

1) Download resource NLTK
- gutenberg: dataset corpus Gutenberg
- punkt / punkt_tab: model tokenisasi untuk word_tokenize

```python
nltk.download('gutenberg')
nltk.download('punkt_tab')
nltk.download('punkt')
```

2) Load sample text dari Gutenberg corpus (Ambil 1000 karakter pertama dari file austen-emma.txt)

```python
text = nltk.corpus.gutenberg.raw('austen-emma.txt')[:1000]

print("Sample text (first 200 chars):")
print(text[:200])
```

3) Tokenisasi teks

```python
tokens = word_tokenize(text)
print("\nFirst 20 tokens:", tokens[:20])
```

4) Statistik dasar token

```python
print("Total tokens:", len(tokens))
print("Unique tokens:", len(set(tokens)))
```

5) Analisis sederhana: 10 token paling sering muncul

```python
freq = Counter(tokens)
print("\nTop 10 most common tokens:", freq.most_common(10))
```

```python
with open("gutenberg_tokens.txt", "w", encoding="utf-8") as f:
    f.write("\n".join(tokens))

print("\nSaved tokens to: gutenberg_tokens.txt")
```

```python
from datasets import load_dataset
from nltk.tokenize import word_tokenize
from collections import Counter

nltk.download("punkt")
```

```python
print("\nIndoNLU SMSA (Indonesia)")
ds = load_dataset("indonlp/indonlu", "smsa", split="train")

print("Columns:", ds.column_names)
print("Example row:", ds[0])
```

```python
# ambil 3 contoh
sample = ds.select(range(3))
all_tokens_lower = []

for i in range(3):
    text = sample[i]["text"]
    label = sample[i]["label"]

    tokens = word_tokenize(text)

    print(f"\nContoh {i+1} (Label: {label})")
    print("Teks:", text)
    print("20 token pertama:", tokens[:20])
    print("Total token:", len(tokens))
    print("Token unik:", len(set(tokens)))

    #simpen token
    filename = f"smsa_tokens_sample_{i+1}.txt"
    with open(filename, "w", encoding="utf-8") as f:
        f.write("\n".join(tokens))
    print(f"Saved: {filename}")
```

##Nyoba Stemming

```python
!pip -q install Sastrawi

from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from nltk.tokenize import word_tokenize

print("\nIndoNLU SMSA — STEMMING")

stemmer = StemmerFactory().create_stemmer()

for i in range(3):
    text = sample[i]["text"]
    label = sample[i]["label"]

    stemmed_text = stemmer.stem(text)
    stemmed_tokens = word_tokenize(stemmed_text)

    print(f"\nContoh {i+1} (Label: {label})")
    print("Teks setelah stemming:", stemmed_text)
    print("Token setelah stemming (first 20):", stemmed_tokens[:20])
    print("Jumlah token setelah stemming:", len(stemmed_tokens))

    outname = f"smsa_stemmed_tokens_sample_{i+1}.txt"
    with open(outname, "w", encoding="utf-8") as f:
        f.write("\n".join(stemmed_tokens))
    print(f"Saved: {outname}")
```
