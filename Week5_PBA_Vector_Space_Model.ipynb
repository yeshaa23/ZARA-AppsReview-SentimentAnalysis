```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
```

```python
# 1. Dataset dari slide Stanford
corpus = [
    "dalam setahun terakhir tiga kilang minyak milik pertamina kebakaran",
    "dpr mendesak pt pertamina melakukan evaluasi mendalam terkait kebakaran kilang minyak",
    "pertamina harus memastikan keamanan dan keselamatan masyarakat agar kebakaran tidak terulang kembali",
    "pertamina diminta meningkatkan sistem pengamanan kilang minyak"
]

# 2. Representasi Vektor (Hanya fokus pada kata 'hate' dan 'darkness')
vectorizer = CountVectorizer(vocabulary=['pertamina', 'kebakaran'])
X = vectorizer.fit_transform(corpus)

```

```python
# 3. Hitung Cosine Similarity
# Membandingkan semua dokumen dengan Dokumen 0 (1 John)
cos_sim = cosine_similarity(X[0:1], X)

# Tampilkan Hasil
df = pd.DataFrame(X.toarray(), columns=['pertamina', 'kebakaran'])
df['Cosine Sim to Doc 0'] = cos_sim.flatten()
print(df)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# 1. Dataset dari artikel
corpus = [
    "kilang minyak milik Pertamina mengalami kebakaran",
    "dpr mendesak Pertamina melakukan evaluasi atas kebakaran kilang",
    "pertamina harus memastikan keamanan masyarakat di sekitar kilang",
    "dpr menyoroti pentingnya keselamatan masyarakat di sekitar industri energi"
]

# 2. Representasi Vektor
vectorizer = CountVectorizer(vocabulary=['pertamina', 'kebakaran'])
X = vectorizer.fit_transform(corpus)

# 3. Cosine similarity
cos_sim = cosine_similarity(X[0:1], X)

# 4. Tampilkan hasil
df = pd.DataFrame(X.toarray(), columns=['pertamina', 'kebakaran'])
df['Cosine Sim to Doc 0'] = cos_sim.flatten()

print(df)
```

**CountVectorizer:** Mengubah teks menjadi angka (matriks frekuensi kata). Di sini kita membatasi fitur hanya pada kata "hate" dan "darkness" agar sesuai dengan grafik di slide Stanford.

**Cosine Similarity vs Distance:** Fungsi ini menghasilkan nilai 0 sampai 1. Semakin mendekati 1, semakin mirip kedua dokumen tersebut.

**Analisis Hasil:** Bahwa Dokumen 2 (MLK) memiliki nilai similarity yang tinggi terhadap Dokumen 0, membuktikan bahwa meskipun jumlah katanya jauh lebih banyak, "arah" topiknya tetap serupa.

**Euclidean Distance:** Jika kita menggunakan jarak lurus, Dokumen 2 akan dianggap sangat berbeda karena jumlah kemunculan katanya (3 dan 4) jauh dari Dokumen 0 (1 dan 2).

**Aplikasi di Dunia Nyata:** Teknik ini adalah dasar dari mesin pencari (search engines) dan sistem rekomendasi teks yang sering kita gunakan sehari-hari.

```python
import numpy as np
def length(v):
    return np.sqrt((v ** 2).sum())
def cos_dist(v, w):
    return 1 - (v * w).sum() / (length(v) * length(w))
cos_dist(tf.loc[0], tf.loc[1]), cos_dist(tf.loc[0], tf.loc[2])
```
